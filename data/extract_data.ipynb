```python
def get_results(album_id,sp):
    results = sp.playlist(album_id)
    return results

def get_dataframe_from_album(results, sp, next = False):
    
    if next == False:
        results_tracks = results['tracks']
    
        songs = results_tracks['items']
    else:
        songs = results['items']

    list_id = []
    list_name = []
    list_artist = []
    list_disc = []
    list_duration = []
    list_episode = []
    list_explicit = []
    list_popularity = []
    list_danceability = []
    list_acousticness = []
    list_energy = [] 
    list_instrumentalness = []
    list_liveness = []
    list_loudness = []
    list_mode = []
    list_speechiness = []
    list_tempo = []
    list_time_signature = []
    list_valence = []
    if next == False:
        for i in songs:
            audio_features = sp.audio_features(i['track']['id'])[0]
            list_id.append(i['track']['id'])
            list_name.append(i['track']['name'])
            list_artist.append(i['track']['artists'][0]['name'])
            list_disc.append(i['track']['disc_number'])
            list_duration.append(i['track']['duration_ms'])
            list_episode.append(i['track']['episode'])
            list_explicit.append(i['track']['explicit'])
            list_popularity.append(i['track']['popularity'])
            list_danceability.append(audio_features['danceability'])
            list_acousticness.append(audio_features['acousticness'])
            list_energy.append(audio_features['energy'])
            list_instrumentalness.append(audio_features['instrumentalness'])
            list_liveness.append(audio_features['liveness'])
            list_loudness.append(audio_features['loudness'])
            list_mode.append(audio_features['mode'])
            list_speechiness.append(audio_features['speechiness'])
            list_tempo.append(audio_features['tempo'])
            list_time_signature.append(audio_features['time_signature'])
            list_valence.append(audio_features['valence'])
    else:
        for i in songs:
            audio_features = sp.audio_features(i['id'])[0]
            list_id.append(i['id'])
            list_name.append(i['name'])
            list_artist.append(i['artists'][0]['name'])
            list_disc.append(i['disc_number'])
            list_duration.append(i['duration_ms'])
            list_episode.append(i['episode'])
            list_explicit.append(i['explicit'])
            list_popularity.append(i['popularity'])
            list_danceability.append(audio_features['danceability'])
            list_acousticness.append(audio_features['acousticness'])
            list_energy.append(audio_features['energy'])
            list_instrumentalness.append(audio_features['instrumentalness'])
            list_liveness.append(audio_features['liveness'])
            list_loudness.append(audio_features['loudness'])
            list_mode.append(audio_features['mode'])
            list_speechiness.append(audio_features['speechiness'])
            list_tempo.append(audio_features['tempo'])
            list_time_signature.append(audio_features['time_signature'])
            list_valence.append(audio_features['valence'])

    data = {'id':list_id,
    'name':list_name,
    'artist':list_artist,
    'disc':list_disc,
    'duration':list_duration,
    'episode':list_episode,
    'explicit':list_explicit,
    'popularity':list_popularity,
    'danceability':list_danceability,
    'acousticness':list_acousticness,
    'energy':list_energy,
    'instrumentalness':list_instrumentalness,
    'liveness':list_liveness,
    'loudness':list_loudness,
    'mode':list_mode,
    'speechiness':list_speechiness,
    'tempo':list_tempo,
    'time_signature':list_time_signature,
    'valence':list_valence}

    dataset = pd.DataFrame(data)

    return dataset,results
```
